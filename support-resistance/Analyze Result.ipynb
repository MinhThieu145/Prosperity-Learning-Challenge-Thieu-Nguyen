```python
import mplfinance as mpf
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import pandas as pd
```

```python
# read csv, set sepereator, set index to day
df = pd.read_csv('test-1-dataset.csv', sep=';',  parse_dates=True)
df.head()
```

```python
fig, ax = plt.subplots()

for i, row in df.iterrows():
    ax.scatter([row['bid_price_1'], row['bid_price_2'], row['bid_price_3']], 
               [row['bid_volume_1'], row['bid_volume_2'], row['bid_volume_3']], 
               color='green', label='Bids' if i == 0 else "")
    ax.scatter([row['ask_price_1'], row['ask_price_2'], row['ask_price_3']], 
               [row['ask_volume_1'], row['ask_volume_2'], row['ask_volume_3']], 
               color='red', label='Asks' if i == 0 else "")

plt.title('Bid and Ask Prices with Volumes')
plt.xlabel('Price')
plt.ylabel('Volume')
plt.legend()
plt.show()
```
