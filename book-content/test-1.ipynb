### Import data

```python
import yfinance as yahooFinance

import pandas as pd
```

```python
GetFacebookInformation = yahooFinance.Ticker("META")


df = (GetFacebookInformation.history(period="1y"))

df
```

```python
# flat the dataframe
df.reset_index(inplace=True)
df
```

```python

import mplfinance as mpf


# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Set the 'Date' column as the index
df.set_index('Date', inplace=True)

# Plotting the candlestick chart
mpf.plot(df, type='candle', style='charles', volume=False, title="Sample Candlestick Chart")
```

```python
# get the date

meta_df = pd.DataFrame(index=df.index)
meta_df
```

```python
# get the price
meta_df['price'] = df['Close']

# daily difference
meta_df['daily_difference'] = df.diff()['Close']
```

```python
# flat the dataframe
meta_df = meta_df.reset_index()
meta_df
```

```python
meta_df['signal'] = 0
meta_df['signal'] = meta_df['daily_difference'].apply(lambda x: 1 if x > 0 else 0)

meta_df
```

```python
# get position for each day
meta_df['position'] = meta_df['signal'].diff()
meta_df
```

### Signal Visualization

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax1 = fig.add_subplot(111, ylabel='Price in $')

# plot
meta_df['price'].plot(ax=ax1, color='r', lw=2.)

# add up arrows when we buy
ax1.plot(meta_df.loc[meta_df.position == 1].index,
         meta_df.price[meta_df.position == 1],
         '^', markersize=5, color='m')

# add down arrows when we sell
ax1.plot(meta_df.loc[meta_df.position == -1].index,
         meta_df.price[meta_df.position == -1],
         'v', markersize=5, color='k')

```
